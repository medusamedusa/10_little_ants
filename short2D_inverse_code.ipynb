```python
import numpy as np
from numpy import pi
import random
import matplotlib.pyplot as plt
from numpy import array
from scipy import newaxis
from pydub import AudioSegment
from pydub.playback import play
```

```python
class RoboSound(object):
    _registry = []

    def __init__(self, name, note, betax, betay):
        self._registry.append(self)
        self.name = name
        self.note = note
        self.betax = betax
        self.betay = betay
```

```python
R1 = RoboSound('R1', None, None, None)
R2 = RoboSound('R2', None, None, None)
R3 = RoboSound('R3', None, None, None)
R4 = RoboSound('R4', None, None, None)
R5 = RoboSound('R5', None, None, None)
R6 = RoboSound('R6', None, None, None)
R7 = RoboSound('R7', None, None, None)
R8 = RoboSound('R8', None, None, None)
R9 = RoboSound('R9', None, None, None)
R10 = RoboSound('R10', None, None, None)
```

```python
# User input of 10 notes!!! (at each time point...)

for i in RoboSound._registry:
    i.note = input("Enter a note:")

for i in RoboSound._registry:
    print(i.name, i.note)
```

```python
# setting up positions and create the sound file


audio = []

for x in range(11): # it should be between 1 and 11 in this case
    valuez = AudioSegment.from_file("notes_/tC.mp3") # just to initialize
    audio.append(valuez)

for i in RoboSound._registry:
    if (i.note == 'C'):
        i.betax = 0
        i.betay = 0.5
        valuez = AudioSegment.from_file("notes_/tC.mp3")
        audio.append(valuez)
    if (i.note == 'B'):
        i.betax = round(np.random.uniform(0,0.17), 3)
        i.betay = round(np.random.uniform(0,0.5), 3)
        valuez = AudioSegment.from_file("notes_/tB.mp3")
        audio.append(valuez)
    if (i.note == 'C#'):
        i.betax = round(np.random.uniform(0.17, 0.3), 3)
        i.betay = round(np.random.uniform(0.5, 1), 3)
        valuez = AudioSegment.from_file("notes_/tC#.mp3")
        audio.append(valuez)
    if (i.note == 'D'):
        i.betax = round(np.random.uniform(0.17, 0.3), 3)
        i.betay = round(np.random.uniform(0.5, 1), 3)
        valuez = AudioSegment.from_file("notes_/tD.mp3")
        audio.append(valuez)
    if (i.note == 'D#'):
        i.betax = round(np.random.uniform(0.3, 0.5), 3)
        i.betay = round(np.random.uniform(0, 0.5), 3)
        valuez = AudioSegment.from_file("notes_/tD#.mp3")
        audio.append(valuez)
    if (i.note == 'E'):
        i.betax = round(np.random.uniform(0.5, 0.64), 3)
        i.betay = round(np.random.uniform(0, 0.5), 3)
        valuez = AudioSegment.from_file("notes_/tE.mp3")
        audio.append(valuez)
    if (i.note == 'F'):
        i.betax = round(np.random.uniform(0.64, 0.84), 3)
        i.betay = round(np.random.uniform(0,0.5), 3)
        valuez = AudioSegment.from_file("notes_/tF.mp3")
        audio.append(valuez)
    if (i.note == 'F#'):
        i.betax = round(np.random.uniform(0.84, 1), 3)
        i.betay = 0.5 # ok?
        valuez = AudioSegment.from_file("notes_/tF#.mp3")
        audio.append(valuez)
    if (i.note == 'G'):
        i.betax = round(np.random.uniform(0.64, 0.84), 3)
        i.betay = round(np.random.uniform(0.5, 1), 3)
        valuez = AudioSegment.from_file("notes_/tG.mp3")
        audio.append(valuez)
    if (i.note == 'G#'):
        i.betax = round(np.random.uniform(0.5, 0.64), 3)
        i.betay = round(np.random.uniform(0.5, 1), 3)
        valuez = AudioSegment.from_file("notes_/tG#.mp3")
        audio.append(valuez)
    if (i.note == 'A#'):
        i.betax = round(np.random.uniform(0.3, 0.5), 3)
        i.betay = round(np.random.uniform(0.5, 1), 3)
        valuez = AudioSegment.from_file("notes_/tA#.mp3")
        audio.append(valuez)
    if (i.note == 'B'):
        i.betax = round(np.random.uniform(0.17, 0.3), 3)
        i.betay = round(np.random.uniform(0, 0.5), 3)
        valuez = AudioSegment.from_file("notes_/tB.mp3")
        audio.append(valuez)



mix = []

for s in range(11): # it should be between 1 and 11 
    values = audio[s].overlay(audio[s+1])
    values2 = values.overlay(audio[s+2])
    values3 = values2.overlay(audio[s+3])
    values4 = values3.overlay(audio[s+4])
    values5 = values4.overlay(audio[s+5])
    values6 = values5.overlay(audio[s+6])
    values7 = values6.overlay(audio[s+7])
    values8 = values7.overlay(audio[s+8])
    values9 = values8.overlay(audio[s+9])
    mix.append(values9)
    print(mix[s])

mix[10].export("notes_/10_robot_sound/mixed_time_inverse_method.mp3", format='mp3') # export mixed  audio file
play(mix[10]) 
```
